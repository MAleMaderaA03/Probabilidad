```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()

# Cargar los datos (suponiendo que están en un archivo CSV)
df = pd.read_excel("Natalidad_02.xlsx")

# Convertir la columna 'nacimientos' a numérica,
# manejando errores con 'coerce' para convertir valores no numéricos a NaN
df['nacimientos'] = pd.to_numeric(df['nacimientos'], errors='coerce')
df['Año'] = pd.to_numeric(df['Año'], errors='coerce')

# Calcular medidas de tendencia central y dispersión
media = df['nacimientos'].mean()
mediana = df['nacimientos'].median()
moda = df['nacimientos'].mode()[0]
desviacion_estandar = df['nacimientos'].std()
coeficiente_variacion = desviacion_estandar / media

# Crear el histograma
plt.figure(figsize=(15, 6))
plt.hist(df['Año'], bins=10, weights=df["nacimientos"])
plt.xlabel('Año')
plt.ylabel('Numero de Nacimientos')
plt.title('Distribución de Nacimientos en Durango')
plt.show()

# Imprimir los resultados
print('Media:', media)
print('Mediana:', mediana)
print('Moda:', moda)
print('Desviación Estándar:', desviacion_estandar)
print('Coeficiente de Variación:', coeficiente_variacion)
```
